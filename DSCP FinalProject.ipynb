# Salary Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Salary_Data.csv')
df["Education Level"] = df["Education Level"].apply(lambda x:"Master's Degree" if x=="Master's" else x)
df["Education Level"] = df["Education Level"].apply(lambda x:"Bachelor's Degree" if x=="Bachelor's" else x)
df["Education Level"] = df["Education Level"].apply(lambda x:"PhD" if x=="phD" else x)
```

```python
fig, ax = plt.subplots()

ax.set_title("Salary by age differences")
ax.set_xlabel("Age")
ax.set_ylabel("Salaries ($)")

X = df["Age"]
Y = df["Salary"]

ax.scatter(X,Y,s=12.5)
```

```python
fig, ax = plt.subplots()

ax.set_title("Salary by Year of Experience differences")
ax.set_xlabel("Years of Experience")
ax.set_ylabel("Salaries ($)")

X = df["Years of Experience"]
Y = df["Salary"]

ax.scatter(X,Y,s=12.5)
```

```python
fig, ax = plt.subplots()

X = ['Female', 'Male']
Y = [107888.9986, 121389.8709]

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center')

ax.set_title("Average Salary by Gender differences")
ax.set_xlabel("Gender")
ax.set_ylabel("Average Salaries ($)")

bar_labels = ['#e75480' , '#6CA0DC']

ax.bar(X,Y, color=bar_labels)
addlabels(X,Y)
```
